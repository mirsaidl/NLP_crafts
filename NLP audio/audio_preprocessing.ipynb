```python
import librosa
import pandas as pd
import matplotlib.pyplot as plt
import IPython.display as ipd
import librosa.display
```

```python
# Load the audio file
audio_file = 'audio sample/audio.mp3'
data, rate = librosa.load(audio_file)
data, rate
```

```python
data.shape
```

```python
ipd.Audio(audio_file, rate=rate)
```

```python
pd.Series(data).plot(figsize=(10,5),
                     lw=1,
                     title='Raw Audio Example')

plt.show()
```

```python
melspec = librosa.feature.melspectrogram(y=data, sr=rate)
```

```python
log_s = librosa.amplitude_to_db(melspec)
```

```python
plt.figure(figsize=(10,5))
librosa.display.specshow(log_s, sr=rate, x_axis='time', y_axis='mel')
plt.show()
```
